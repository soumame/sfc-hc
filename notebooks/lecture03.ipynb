## JOINについて

2つのテーブルを結合できる。

結合方法には、
- inner join
- outer join (left or right)
があるが、デフォルトではinner joinである


```
SELECT
    ...
FROM
    tableA
JOIN
    tableB
ON
    tableA.column = tableB.column
WHERE

```

## WITH文


一時テーブルを使って処理する

```
WITH query_name1 as (
SELECT
...
),
query_name2 as (
....
), ....

SELECT
....
```

````python
# 一人のユーザーが何回記事をクリックしたか
sql="""
    SELECT
        user_id,
        COUNT(article_id) as num_clicks
    FROM
        clicks
    GROUP BY user_id
"""

pandas.read_sql(sql, sqlite_conn)
````

- 記事を１回しかクリックしてない人は、何人くらいか？
- 記事を２回クリックした人は、何人くらいか？
- このように、分布を作りたい場合は...(ヒストグラムを作りたい)

FROMの中にテーブルを突っ込むことができる。

````python
# クリック数の分布

sql = """
SELECT
    num_clicks,
    COUNT(user_id) as count
FROM (
    SELECT
        user_id,
        COUNT(article_id) as num_clicks
    FROM
        clicks
    GROUP BY user_id
)
GROUP BY 
    num_clicks
"""
print(pandas.read_sql(sql, sqlite_conn))

pandas.read_sql(sql, sqlite_conn).plot(kind='bar', x='num_clicks', y='count')
````

でもこれを長いSQLで書いちゃうと読みづらいので、WITH文をつかって書き換える

````python
# WITH文をつかって書き換える
# ユーザのクリック数の分布
# 結果は同じようになる
sql = """
WITH user_num_clicks as (
    SELECT
        user_id,
        COUNT(article_id) as num_clicks
    FROM
        clicks
    GROUP BY user_id
)
SELECT
    num_clicks,
    COUNT(user_id)
FROM 
    user_num_clicks
GROUP BY 
    num_clicks
"""
pandas.read_sql(sql, sqlite_conn)
````


説明を書いておく

```sql
WITH article_num_clicks as (
    SELECT
        article_id,
        COUNT(user_id) as num_clicks
    FROM
        clicks
    GROUP BY article_id
)
```
ここで、article_num_clicksという一時テーブルを作成している。各記事のクリック回数を並べる
- 1番の記事はx回クリックされている
- 2番の記事はy回クリックされている
- ...

そして、この一時テーブルを使って、記事のクリック数の分布を計算する

```sql
SELECT
    num_clicks,
    COUNT(article_id) as num_articles
FROM
    article_num_clicks
GROUP BY
    num_clicks
```
COUNT(article_id)は、article_num_clicksテーブルの中で、num_clicksがx回の記事の数を数えている。
- クリック数が1回の記事は、何記事あるか？
- クリック数が2回の記事は、何記事あるか？
- ...

というように、記事のクリック数の分布を計算することができる。



````python
# WITH文を使って、記事のクリック数の分布を計算してみよう
sql = """
WITH article_num_clicks as (
    SELECT
        article_id,
        COUNT(user_id) as num_clicks
    FROM
        clicks
    GROUP BY article_id
)
SELECT
    num_clicks,
    COUNT(article_id) as num_articles
FROM
    article_num_clicks
GROUP BY
    num_clicks
"""

print(pandas.read_sql(sql, sqlite_conn))

pandas.read_sql(sql, sqlite_conn).plot(kind='bar', x='num_clicks', y='num_articles')
````

````python
# WITHとJOINを組み合わせる
# click_uuとclick_numを計算し、ひとりあたりのクリック数の推移をみる

sql = """
WITH click_uu as (
    SELECT
        DATE(created_at) as date,
        COUNT(DISTINCT user_id) as uu
    FROM
        clicks
    GROUP BY
        DATE(created_at)
),
click_num as (
    SELECT
        DATE(created_at) as date,
        COUNT(user_id) as num
    FROM
        clicks
    GROUP BY
        DATE(created_at) 
)
SELECT
    click_num.date,
    click_uu.uu,
    click_num.num,
    1.0 * click_num.num / click_uu.uu as click_per_uu
FROM
    click_num
JOIN
    click_uu
ON
    click_num.date = click_uu.date
"""

pandas.read_sql(sql, sqlite_conn)
````

````python
# 1記事のクリック数の推移を計算しよう
# この日は１記事あたり平均何回クリックされたか
sql = """
WITH click_num_by_article as (
    SELECT
        DATE(created_at) as date,
        article_id,
        COUNT(user_id) as click_num
    FROM
        clicks
    GROUP BY
        DATE(created_at),
        article_id
)
SELECT
    click_num_by_article.date,
    COUNT(click_num_by_article.article_id) as num_articles,
    SUM(click_num_by_article.click_num) as click_num,
    1.0 * SUM(click_num_by_article.click_num) / COUNT(click_num_by_article.article_id) as click_per_article
FROM
    click_num_by_article
GROUP BY
    click_num_by_article.date
"""

pandas.read_sql(sql, sqlite_conn)
````

````python
# 登録日から継続日数ごとのクリック数

sql = """
WITH click_num as (
    SELECT
        DATE(created_at) as date,
        user_id,
        COUNT(article_id) as click_num
    FROM
        clicks
    GROUP BY
        DATE(created_at),
        user_id
),
click_num_days as (
    SELECT
        JULIANDAY(click_num.date) - JULIANDAY(DATE(users.created_at)) as conitnue_days,
        click_num.user_id,
        click_num.click_num as click_num
    FROM
        click_num
    JOIN
        users
    ON
        users.user_id = click_num.user_id
)
SELECT
    conitnue_days,
    SUM(click_num) as click_num,
    COUNT(user_id) as click_uu,
    1.0 * SUM(click_num) / COUNT(DISTINCT user_id) as click_num_per_uu
FROM
    click_num_days
GROUP BY
    conitnue_days
"""

pandas.read_sql(sql, sqlite_conn)
````
